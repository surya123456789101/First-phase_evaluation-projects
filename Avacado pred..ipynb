```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
Avacado_dataset = pd.read_csv('avocado.csv')
```

```python
Avacado_dataset.replace({"type":{'conventional':0,'organic':1}},inplace=True)
```

```python
Avacado_dataset.head()
```

```python
Avacado_dataset['Date']= pd.to_datetime(Avacado_dataset['Date'], format="%Y-%m-%d")
```

```python
Avacado_dataset.info()
```

```python
Avacado_dataset.select_dtypes(include="float64").columns
```

```python
Avacado_dataset = Avacado_dataset.astype({"Total Volume":'str',"4046": 'str',"4225":'str',"4770":'str',"Total Bags":'str',
                                         "Small Bags":'str',"Large Bags":'str',"XLarge Bags":'str',"AveragePrice":'str'})

print(Avacado_dataset.dtypes)
```

```python
Avacado_dataset['month']=pd.DatetimeIndex(Avacado_dataset.Date).month
Avacado_dataset.drop('Date',axis=1,inplace=True)
```

```python
Avacado_dataset.head()
```

```python
print(Avacado_dataset.dtypes)
```

```python
X = Avacado_dataset.drop('region',axis=1)
Y = Avacado_dataset['region']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, Y_train)
```

```python
X_test_prediction = model.predict(X_test)

data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy:', data_accuracy)
```
